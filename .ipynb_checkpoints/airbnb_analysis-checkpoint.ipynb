```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Load and preview the Airbnb listings raw data
listings_df = pd.read_csv('listings-summary.csv')
listings_df.head()
```

```python
#Select relevant columns to keep in the dataframe
cleaned_reviews_df = listings_df[['neighbourhood', 'number_of_reviews']]
cleaned_reviews_df.head()
```

```python
grouped_df = cleaned_reviews_df.groupby(cleaned_reviews_df['neighbourhood']).sum().reset_index()
grouped_df = grouped_df['number_of_reviews']/grouped_df['number_of_reviews'].sum() * 1000
```

```python
grouped_df.describe() 
```

```python
plt.hist(grouped_df['number_of_reviews'], color='blue', edgecolor='black', bins= int(10000/250))
plt.show()
```
